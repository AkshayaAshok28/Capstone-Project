```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_financials_main=pd.read_csv("/content/Finance_main.csv")
df_financials_main
```

```python
df_finance_details=pd.read_csv("/content/Finance_details.csv")
df_finance_details
```

```python
 #YEAR WISE LOAN AMOUNT STATS: PANDAS
 df_financials_main['loan_amnt'] = pd.to_numeric(df_financials_main['loan_amnt'], errors='coerce')
 df_financials_main['issue_d']=pd.to_datetime(df_financials_main['issue_d'],errors='coerce')
 df_financials_main['issue_year'] = df_financials_main['issue_d'].dt.year
 df_financials_main['issue_year'] = df_financials_main['issue_year'].astype(int)
 Yearly_stats = df_financials_main.groupby('issue_year')['loan_amnt'].agg(Total_Loan_Amount='sum').reset_index()
 Yearly_stats
```

```python
#SEABORN:
sns.barplot(x='issue_year',y='Total_Loan_Amount',data=Yearly_stats, palette='viridis')
plt.xlabel('Year')
plt.ylabel('Total Loan Amount')
plt.title('Total Loan Amount Disbursed Per Year', fontsize=16)
plt.show()
```

```python
#MATPLOTLIB
plt.bar(Yearly_stats['issue_year'],Yearly_stats['Total_Loan_Amount'])
plt.xlabel('Year')
plt.ylabel('Total Loan Amount')
plt.title('Total loan amount by Year')
plt.show()
```

```python
#Grade and Subgrade wise Revolve Balance
df_finance_details['revol_bal']=pd.to_numeric(df_finance_details['revol_bal'],errors='coerce')
df_financials_main['grade']=df_financials_main['grade'].astype(str)
df_financials_main['sub_grade']=df_financials_main['sub_grade'].astype(str)
df_financials_main['grade_sub_grade']=df_financials_main['grade']+ df_financials_main['sub_grade']
left_join=pd.merge(df_finance_details,df_financials_main,on='id',how='left')
left_join.groupby('grade_sub_grade')['revol_bal'].agg(Total_Revolving_Balance='sum').reset_index()

```

```python
#Total payment of verified status and Total payment of non verified status
df_finance_details['total_pymnt']=pd.to_numeric(df_finance_details['total_pymnt'],errors='coerce')
df_financials_main['verification_status']=df_financials_main['verification_status'].astype(str)
left_join=pd.merge(df_finance_details,df_financials_main,on='id',how='left')
lj=left_join.groupby('verification_status')['total_pymnt'].agg(Total_payment_of_verified_status_and_non_verified_status='sum').reset_index()
lj
```

```python
status_counts=left_join['verification_status'].value_counts()
labels=status_counts.index
sizes=status_counts.values
colors=['teal','darkorange']
plt.pie(sizes, labels=labels,colors=colors, autopct='%1.1f%%', explode=(0.05, 0), startangle=90,shadow=True,
        textprops={'fontsize': 10, 'color': 'black'},radius=1.1)
```

```python
#STATE WISE MONTH WISE LOAN STATUS
df_financials_main['addr_state']=df_financials_main['addr_state'].astype(str)
df_financials_main['issue_d']=pd.to_datetime(df_financials_main['issue_d'],errors='coerce')
df_financials_main['issue_month_name'] = df_financials_main['issue_d'].dt.month_name()
df_financials_main['issue_month_name'] = df_financials_main['issue_month_name'].astype(str)
State_month_stats=df_financials_main.groupby(['addr_state','issue_month_name'])['loan_status'].value_counts().reset_index(name='count')
State_month_stats
```

```python
#matplotlib for the above data
#suggest me a visual in matplotlib for the above context having 2 categorical data
plt.figure(figsize=(10,6))
sns.barplot(x='addr_state',y='count',hue='issue_month_name',data=State_month_stats,palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('count')
plt.title('Loan status by state and month')
plt.show()


```

```python
#Home ownership vs last payment date stats
df_financials_main['home_ownership']=df_financials_main['home_ownership'].astype(str)
df_finance_details['last_pymnt_d']=pd.to_datetime(df_finance_details['last_pymnt_d'],errors='coerce')
left_join=pd.merge(df_finance_details,df_financials_main,how='left',on='id')
last_payment_stats=left_join.groupby('home_ownership')['last_pymnt_d'].agg(last_payment_stats='count').reset_index()
last_payment_stats
```

```python
#matplotlib chart for the above context
plt.bar(last_payment_stats['home_ownership'],last_payment_stats['last_payment_stats'],color='green')
plt.xlabel('Home Ownership')
plt.ylabel('Last Payment Date')
plt.title('Last Payment Date by Home Ownership')
plt.show()
```
